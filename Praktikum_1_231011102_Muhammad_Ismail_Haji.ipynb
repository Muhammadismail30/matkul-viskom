**Tugas No. 2 Melakukan ekstraksi**

Buatlah suatu kode python yang dapat melakukan ekstraksi warna kuning pada bunga tulip dengan memanfaatkan format warna HSV.

```python
from google.colab.patches import cv2_imshow
import cv2
```

Mengimpor tools yang akan digunakan untuk pemrosesan citra pada gambar.

```python
img = cv2.imread('tulips.jpg')
```

Membaca file gambar 'tulips.jpg' yang diinisialisasi pada variabel img.

```python
plt.imshow(img[:,:,::-1])
```

Menampilkan gambar asli dengan format RGB.

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

***cv2.cvtColor*** berfungsi mengonversi gambar dari RGB menjadi HSV (Hue, Saturation, Value). Ini bertujuan untuk segmentasi warna.

```python
lr = np.array([20, 100, 100]) #batas bawah warna kuning
ur = np.array([40, 255, 255]) #batas atas warna kuning
mask = cv2.inRange(hsv, lr, ur)
```

Menentukan rentang HSV untuk menangkap warna kuning.

```python
type(mask)
```

```python
mask.dtype
```

```python
mask.shape
```

```python
mask.min()
```

```python
np.unique(mask)
```

```python
res = cv2.bitwise_and(img, img, mask=mask)
```

Mengisolasi warna kuning dan menyaring warna lain.

**Membandingkan gambar asli dengan hasil ekstraksi.**

```python
plt.figure(1)
plt.subplot(121)
plt.imshow(img[:,:,::-1])
plt.subplot(122)
plt.imshow(res[:,:,::-1])
plt.show()
```
